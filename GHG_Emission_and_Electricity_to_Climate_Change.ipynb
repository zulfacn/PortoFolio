Climate change has been problem to all countries around the world. Since 20th century, green house gas emission (GHG Emission) which is the result from human activities has became the main reason of climate change. There is no way we would stop activities that resulting GHG Emission, the only way is to reduce it. There are 2 areas that need more attention in producing GHG Emission, energy and food/agriculture. Therefore, we will conduct analysis to find correlation between the increasing of earth temperature with GHG Emission level. Then we will find which factors give highest contribution to GHG Emission.

### **3. Pre-Processing Data; Check for Missing Value and Replace Missing Value**

Delete unused data and make the year only 2000 to 2016

```
thn= [i for i in range(1990,2000)]
for i, val in enumerate(emisi['Year'].values):
    if val in thn:
        emisi.drop(index=i, axis=0, inplace=True,
    errors='ignore')
    else:
            continue
```

**Clustering GHG Emission**

```
#normalizing based on standard deviason
from sklearn.preprocessing import StandardScaler
#X7 = df_electricityd.values[:,0:]
X7 = df_ghgd['ghg-emission']
X7 = np.nan_to_num(X7).reshape(-1,1)
Clus_dataSet = StandardScaler().fit_transform(X7)
Clus_dataSet
```

```
#K-Mean Clustering
from sklearn.cluster import KMeans
wcss = []
for i in range(1, 11):
    kmeans7 = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
    kmeans7.fit(X7)
    wcss.append(kmeans7.inertia_)
plt.plot(range(1, 11), wcss)
plt.title('Metode Elbow')
plt.xlabel('Jumlah clusters')
plt.ylabel('WCSS')
plt.show()
```

```
clusterNum7 = 3
k_means7 = KMeans(init = "k-means++", n_clusters = clusterNum7, n_init = 15)
k_means7.fit(X7)
labels7 = k_means7.labels_
print(labels7)
```

```
df_ghgd["Cluster"] = labels7
df_ghgd
```

```
# World Map Visualization

import pycountry 
import geopandas
import mapclassify

def alpha3code(column):
    CODE=[]
    for country in column:
        try:
            code=pycountry.countries.get(name=country).alpha_3
           # .alpha_3 means 3-letter country code 
           # .alpha_2 means 2-letter country code
            CODE.append(code)
        except:
            CODE.append('None')
    return CODE
    
# create a column for code 
df_electricityd['Country'] = df_electricity['Country']
df_ghgd['Country'] = df_ghg['Country']

df_electricityd['CODE'] = alpha3code(df_electricity['Country'])
df_ghgd['CODE'] = alpha3code(df_ghg['Country'])
```

```
world = geopandas.read_file(geopandas.datasets.get_path('naturalearth_lowres'))

# rename the columns so that we can merge with our data
world.columns=['pop_est', 'continent', 'name', 'CODE', 'gdp_md_est', 'geometry']
world = world.drop(columns=['pop_est','name','gdp_md_est'])

# then merge with our data 
merge_electricity=pd.merge(world,df_electricityd,on='CODE')
merge_ghg=pd.merge(world,df_ghgd,on='CODE')
```

```
merge_electricity.plot(column='Cluster', scheme="quantiles",
           figsize=(25, 20),
           legend=True ,cmap='BrBG')

plt.title('World Electricity Usage',fontsize=25)
```
